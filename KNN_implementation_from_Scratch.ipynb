```python
import pandas as pd
```

```python
data=pd.read_csv('diamonds (1).csv')
```

```python
data.head()
```

```python
cur_or=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_or=['J','I','H','G','F','E','D']
clarity_or=['I1','SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[cur_or,color_or,clarity_or]).set_output(transform='pandas')
```

```python
oe
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer([('cut_oe',oe,['cut','color','clarity'])],
                     verbose_feature_names_out=False,remainder='passthrough').set_output(transform='pandas')
```

```python
ct
```

```python
data=ct.fit_transform(data)
```

```python
X=data.drop('price',axis=1)
```

```python
y=data[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax=MinMaxScaler().set_output(transform='pandas')
```

```python
ct2=ColumnTransformer([('min_max',minmax,['carat','depth','table','x','y','z'])],
                     verbose_feature_names_out=False,remainder='passthrough').set_output(transform='pandas')
```

```python
x_train_transform=ct2.fit_transform(x_train)
```

```python
x_train_transform
```

```python
x_test_transform=ct2.transform(x_test)
```

```python
x_test_transform
```

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
query=x_test_transform.sample(n=1,random_state=42)
```

```python
query
```

```python
dist=pd.DataFrame(euclidean_distances(x_train_transform,query))
```

```python
top5=dist.sort_values(by=0,ascending=True)
```

```python
top5=top5.iloc[:5,:]
```

```python
index=top5.index
i=list(index)
i
```

```python
knn_manu=y_train.iloc[i,:]
```

```python
knn_manu_price=knn_manu['price'].mean()
```

```python
knn_manu_price #KNN SCRATCH IMPLEMENTATION House price 
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knnr=KNeighborsRegressor(n_neighbors=5)
```

```python
knnr.fit(x_train_transform,y_train)
```

```python
knnr.predict(query)# KNN predicting house price by mode both are same 
```

```python
knnr.score(x_train_transform,y_train)
```

```python
y_pred=knnr.predict(x_test_transform)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
